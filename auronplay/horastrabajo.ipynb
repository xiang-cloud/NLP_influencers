```python
import matplotlib.pyplot as plt

# Datos
actividades = ['Recopilación de datos', 'EDA (Análisis Exploratorio de Datos)', 'Preprocesamiento (Limpieza de datos)', 
               'Minería de Textos', 'Creación de redes', 'Interpretación de resultados', 'Redacción de la memoria']
horas = [20, 11, 40, 22, 33, 25, 40]

# Colores (opcional, pero puede hacer que el gráfico se vea mejor)
colores = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'cyan']

# Crear el gráfico de pastel
plt.figure(figsize=(10, 7))
plt.pie(horas, labels=actividades, colors=colores, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))

# Título
plt.title('Desglosando las horas de trabajo')

# Mostrar el gráfico
plt.show()
```
