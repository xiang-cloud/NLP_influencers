```python
import matplotlib.pyplot as plt

# Datos
actividades = ['Recopilacin de datos', 'EDA (Anlisis Exploratorio de Datos)', 'Preprocesamiento (Limpieza de datos)', 
               'Minera de Textos', 'Creacin de redes', 'Interpretacin de resultados', 'Redaccin de la memoria']
horas = [20, 11, 40, 22, 33, 25, 40]

# Colores (opcional, pero puede hacer que el grfico se vea mejor)
colores = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'cyan']

# Crear el grfico de pastel
plt.figure(figsize=(10, 7))
plt.pie(horas, labels=actividades, colors=colores, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))

# Ttulo
plt.title('Desglosando las horas de trabajo')

# Mostrar el grfico
plt.show()
```
