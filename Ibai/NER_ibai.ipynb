```python
import spacy
from collections import Counter
import pandas as pd

df=pd.read_csv('preprocessed_data.csv')

# Load the Spanish model
nlp = spacy.load('es_core_news_sm')

# Function to perform NER analysis and extract named entities
def perform_ner_analysis(text):
    doc = nlp(text) if isinstance(text, str) else nlp('')
    entities = []
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))
    return entities

# Combine all text columns into a single list
all_text = df['title_preprocessed'].astype(str).tolist() + df['description_preprocessed'].astype(str).tolist() + df['transcript_preprocessed'].astype(str).tolist()

# Perform NER analysis on all text
all_entities = [entity for text in all_text for entity in perform_ner_analysis(text)]

# Count the frequency of each named entity
entity_counter = Counter(all_entities)

# Display the named entities and their frequencies
for entity, count in entity_counter.most_common():
    print(f"Entity: {entity[0]}  Label: {entity[1]}  Frequency: {count}")
```
